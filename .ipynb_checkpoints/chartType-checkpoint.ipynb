```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
import seaborn as sb
sb.set_style('whitegrid')
```

```python
address = "/Users/markchen/Desktop/ExerciseFiles/Data/mtcars.csv"

cars = pd.read_csv(address)
cars.columns = ['carNames', 'mpg', 'cyl', 'disp' , 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.index = cars.carNames
mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.distplot(mpg)
```

```python
cars.plot(kind='scatter', x = 'hp', y = 'mpg', c=['darkgray'], s=150)
```

```python
sb.regplot(x = 'hp', y = 'mpg', data = cars, scatter = True)
```

```python
sb.pairplot(cars)
```

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
plt.show()
```

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sb.boxplot(x='am', y='mpg', data=cars, palette = 'hls')
```
